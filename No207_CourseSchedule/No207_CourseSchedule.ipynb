```python
class Node:
    def __init__(self):
        self.children = None # Children Node. node.children = [node1, node2, ...]
        self.courseNum = None #Course number: 1, 2, etc.
        self.visited = False # True = visited. DFS search temporary status. If visited again, there is a circle. Clear this status after DFS
        self.hasCircleFromMe = False # False, True
        self.checkedFromMe = False # True = checked from this node.
    def __str__(self):
        return f"courseNum: {self.courseNum} visited:{self.visited} hasCircleFromMe:{self.hasCircleFromMe} checkedFromMe:{self.checkedFromMe} children:{self.children}"

nodesDictionary = {}
childrenDictionary = {}

class Solution(object):

    def buildAllNodesDictionary(self, numCourses):
        """
        :type numCourses: int
        :rtype dict: {0: Node0, 1: Node1}
        """
        print("numberCourses " + str(numCourses))

        for x in range(0, numCourses):
            nodesDictionary[x] = Node()
            nodesDictionary[x].courseNum = x
            nodesDictionary[x].visited = False
            nodesDictionary[x].children = []

        return nodesDictionary
    
    def buildChildrenDictionary(self, numCourses, prerequisites):
        """
        :type numCourses: int
        :type prerequisites: List[List[int]]
        :rtype dict
        """

        for x in prerequisites:
        	if x[1] in childrenDictionary:
        		childrenDictionary[x[1]] += [nodesDictionary[x[0]]]
        	else:
        		childrenDictionary[x[1]] = [nodesDictionary[x[0]]]

        return childrenDictionary

    def buildTree(self, numCourses, prerequisites, node):
        """
        :type numCourses: int
        :type prerequisites: List[List[int]]
        :type node: Node
        :type childrenDictionary: dict
        """
        for x in range(0, numCourses):
            nd = nodesDictionary[x]

            if not x in childrenDictionary:
                continue

            if not nd.children:
                nd.children = childrenDictionary[x]
            else:
                nd.children += childrenDictionary[x]
            
    def dfsCheckCircle(self, node):
        print("Checking node" + str(node))
        if node.checkedFromMe:
            print("checkedFromMe true")
            return  
            
        node.checkedFromMe = True
        node.visited = True

        for child in node.children:
            if child.visited:
                node.hasCircleFromMe = True
            else:
                self.dfsCheckCircle(child)
                
        # Clean up temp visited status
        node.visited = False
        
    def dfsAllNodes(self):
        for x in nodesDictionary:
            self.dfsCheckCircle(nodesDictionary[x])

    def hasCircle(self):
        """
        :rtype boolean: true if any of nodes has circle
        """
        for x in nodesDictionary:
            if nodesDictionary[x].hasCircleFromMe:
                return True

        return False

    def canFinish(self, numCourses, prerequisites):
        """
        :type numCourses: int
        :type prerequisites: List[List[int]]
        :rtype: bool
        """
        nodesDictionary.clear()
        childrenDictionary.clear()
        
        if numCourses <= 0:
            return True

        self.buildAllNodesDictionary(numCourses)
        
        self.buildChildrenDictionary(numCourses, prerequisites)

        root = nodesDictionary[0]
        
        self.buildTree(numCourses, prerequisites, root)

        # Print the tree
        self.printTree()

        self.dfsAllNodes()

        return self.hasCircle()

    def printTree(self):
        for x in nodesDictionary:
            print(nodesDictionary[x])
        
s = Solution()

numCourses = 1;
prerequisites = []

b = s.canFinish(numCourses, prerequisites)

print("Has circle: " + str(b))
```
